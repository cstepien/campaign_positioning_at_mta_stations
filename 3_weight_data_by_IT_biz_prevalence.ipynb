```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from copy import deepcopy

%matplotlib inline
```

```python
it_weight = pd.read_csv('zip_of_weights_for_information_workers_2.csv', dtype={'ZIPCODE': str, 'WEIGHT':np.float64})

with open('spatial_4hr_interval_mta.pickle','rb') as read_file:
    time_series_df = pickle.load(read_file)
    
time_series_df = pd.merge(time_series_df, it_weight, on='ZIPCODE', how='left')

# time_series_df.head()
```

```python
time_series_df.ENTRY_DIFF = time_series_df.ENTRY_DIFF*time_series_df.WEIGHT
```

```python
time_series_df.EXIT_DIFF = time_series_df.EXIT_DIFF*time_series_df.WEIGHT
```
